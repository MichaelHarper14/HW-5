```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv

from math import sqrt
from statistics import mean
from scipy.stats import sem
from random import random

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"


# Combine the data into a single dataset


# Display the data table for preview

```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_read_df = pd.read_csv(mouse_drug_data_to_load)
mouse_read_df.head()
```

```python
clinical_trial_data_to_load_df = pd.read_csv(clinical_trial_data_to_load)
clinical_trial_data_to_load_df.head()
```

```python
merge_table = pd.merge(mouse_read_df, clinical_trial_data_to_load_df)
merge_table.head()
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
Mean_Tumor_Volume = merge_table.groupby(["Drug", "Timepoint"])['Tumor Volume (mm3)'].mean()
Mean_Tumor_Volume = pd.DataFrame(Mean_Tumor_Volume)
# Convert to DataFrame
Mean_Tumor_Volume
# Preview DataFrame

```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
SE_Tumor_Volume = merge_table.groupby(["Drug", "Timepoint"])['Tumor Volume (mm3)'].sem()

# Convert to DataFrame
SE_Tumor_Volume = pd.DataFrame(SE_Tumor_Volume)

# Convert to DataFrame
SE_Tumor_Volume.head()
```

```python
Timepoint_Data = Timepoint_Data.set_index('Timepoint')

Timepoint_Data.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked
Tr = Timepoint_Data.pivot_table('Tumor Volume (mm3)', ['Timepoint'], 'Drug')
Tr =pd.DataFrame(Tr)
Tr
```

## 

```python
# Generate the Plot (with Error Bars)

x_axis = np.arange(0,50,5)
tick_locations = [value for value in x_axis]

Capomulin_Tr = Tr["Capomulin"]
Infubinol_Tr = Tr["Infubinol"]
Ketapril_Tr = Tr["Ketapril"]
Placebo_Tr = Tr["Placebo"]

Co = plt.plot(x_axis, Capomulin_Tr, marker = 'o',color = "red", label = 'Capomulin')
Ce = plt.plot(x_axis, Infubinol_Tr, marker = '^',  color= "blue", label='Infubinol')
Cg= plt.plot(x_axis, Ketapril_Tr, marker= 's', color = 'green', label= 'Ketapril')
Cn = plt.plot(x_axis, Placebo_Tr, marker = 'd',color = 'black', label = "Placebo")


plt.errorbar(
[x_axis],
[Capomulin_Tr],
yerr =2,
fmt = 'rs--',
linewidth=1)

plt.errorbar(
[x_axis],
[Infubinol_Tr],
yerr =2,
fmt = 'bs--',
linewidth=1)

plt.errorbar(
[x_axis],
[Ketapril_Tr],
yerr =2,
fmt = 'gs--',
linewidth=1)

plt.errorbar(
[x_axis],
[Placebo_Tr],
yerr =2,

fmt = '--',
color = 'black',
linewidth=1)
plt.legend(loc="best")
plt.xlabel("Treatment Duration (Days)")
plt.title("Tumor Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()

# Save the Figure
plt.savefig("Tr.png")
```
